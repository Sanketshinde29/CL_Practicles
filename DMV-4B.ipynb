```python
# Install pyclustering if not already done
# !pip install pyclustering

import numpy as np
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.cluster import cluster_visualizer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import random

# Suppose your dataset (credit card features) is in a DataFrame called df
# Exclude non-numeric columns
data = df.select_dtypes(include=np.number).values

# Scale the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Try K = 2 to 10 to find the best silhouette score
silhouette_scores = []
for k in range(2, 11):
    # Randomly select initial medoids
    initial_medoids = random.sample(range(len(data_scaled)), k)
    kmedoids_instance = kmedoids(data_scaled, initial_medoids)
    kmedoids_instance.process()
    clusters = kmedoids_instance.get_clusters()
    
    # Flatten cluster labels for silhouette score
    labels = np.zeros(len(data_scaled))
    for cluster_idx, cluster in enumerate(clusters):
        for index in cluster:
            labels[index] = cluster_idx
    
    score = silhouette_score(data_scaled, labels)
    silhouette_scores.append(score)

# Determine optimal K
optimal_k = range(2, 11)[np.argmax(silhouette_scores)]
print("Optimal number of clusters:", optimal_k)

```
